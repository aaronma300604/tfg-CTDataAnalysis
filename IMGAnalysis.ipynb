```python
import numpy as np
import matplotlib.pyplot as plt
import SimpleITK as sitk
import os
```

```python
filename = '/home/roth/aaron/TFG/code/example_img/NRRD/"LUNG1-001"/"LUNG1-001"_CT.nrrd'

lung = sitk.ReadImage(filename)
print(lung.GetSize())
print(type(lung.GetSize()))
print(lung.GetSpacing())
print(lung.GetOrigin())
print(lung.GetDirection())
```

```python
folder = '/home/roth/aaron/TFG/data-corrected-st/NRRD'

patient_size = dict()
patient_spacing = dict()
patient_origin = dict()
patient_direction = dict()

for patient in os.listdir(folder):
    patient_folder = os.path.join(folder, patient)
    if os.path.isdir(patient_folder):
        for file in os.listdir(patient_folder):
            if file.endswith('_CT.nrrd'):
                lung = sitk.ReadImage(filename)
                patient_size[patient] = lung.GetSize()
                patient_spacing[patient] = lung.GetSpacing()
                patient_origin[patient] = lung.GetOrigin()
                patient_direction[patient] = lung.GetDirection()

print(len(patient_size.keys()))
print(len(patient_spacing.keys()))
print(len(patient_origin.keys()))
print(len(patient_direction.keys()))
```

```python
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.hist( [v[0] for v in patient_size.values()])
plt.title('Patient X axis Size Distribution')
plt.xlabel('Size (voxels)')
plt.ylabel('Frequency')

#--------------------------------------
plt.subplot(1, 3, 2)
plt.hist( [v[1] for v in patient_size.values()])
plt.title('Patient Y axis Size Distribution')
plt.xlabel('Size (voxels)')
plt.ylabel('Frequency')

#--------------------------------------
plt.subplot(1, 3, 3)

plt.hist( [v[2] for v in patient_size.values()])
plt.title('Patient Z axis Size Distribution')
plt.xlabel('Size (voxels)')
plt.ylabel('Frequency')

plt.tight_layout()
```

```python
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.hist( [v[0] for v in patient_spacing.values()])
plt.title('Patient X axis Spacing Distribution')
plt.xlabel('Spacing (voxels)')
plt.ylabel('Frequency')

#--------------------------------------
plt.subplot(1, 3, 2)
plt.hist( [v[1] for v in patient_spacing.values()])
plt.title('Patient Y axis Spacing Distribution')
plt.xlabel('Spacing (voxels)')
plt.ylabel('Frequency')

#--------------------------------------
plt.subplot(1, 3, 3)

plt.hist( [v[2] for v in patient_spacing.values()])
plt.title('Patient Z axis Spacing Distribution')
plt.xlabel('Spacing (voxels)')
plt.ylabel('Frequency')

plt.tight_layout()
```

```python
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.hist( [v[0] for v in patient_origin.values()])
plt.title('Patient X axis Origin Distribution')
plt.xlabel('Origin (mm)')
plt.ylabel('Frequency')

#--------------------------------------
plt.subplot(1, 3, 2)
plt.hist( [v[1] for v in patient_origin.values()])
plt.title('Patient Y axis Origin Distribution')
plt.xlabel('Origin (mm)')
plt.ylabel('Frequency')

#--------------------------------------
plt.subplot(1, 3, 3)

plt.hist( [v[2] for v in patient_origin.values()])
plt.title('Patient Z axis Origin Distribution')
plt.xlabel('Origin (mm)')
plt.ylabel('Frequency')

plt.tight_layout()
```

```python
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.hist( [v[0] for v in patient_direction.values()])
plt.title('Patient X axis Direction Distribution')
plt.xlabel('Direction (unit vector)')
plt.ylabel('Frequency')

#--------------------------------------
plt.subplot(1, 3, 2)
plt.hist( [v[1] for v in patient_direction.values()])
plt.title('Patient Y axis Direction Distribution')
plt.xlabel('Direction (unit vector)')
plt.ylabel('Frequency')

#--------------------------------------
plt.subplot(1, 3, 3)

plt.hist( [v[2] for v in patient_direction.values()])
plt.title('Patient Z axis Direction Distribution')
plt.xlabel('Direction (unit vector)')
plt.ylabel('Frequency')

plt.tight_layout()
```
